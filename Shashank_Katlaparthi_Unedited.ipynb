```python
<Title>

</Title>

import os
import random
a = os.popen('ls').read()
a = a.split('\n')
a.remove('')
#a.remove('Untitled.ipynb')
counter = 0
d = {}
segmentuid={}
count = 0
for i in a:
    if 'json' not in i:
        continue
    inp = open(i)
    content = inp.read()
    content = content.lower()
    if 'user not found' in content:
        d[i]= 0
        counter += 1
        #print(content)
    else:
        d[i]= 1
        first=0
        third=0
        value=0
        while ('segmentuid' in content[value:]):
            value += content[value:].find('segmentuid')
            number = int(content[value+13])
            if number==0:
                first += 1
            else:
                third += 1
            value += 1
        segmentuid[i]={'first':first,'third':third}

print(d)
print('\n'+str(counter))
print('\nUser found dictonary 5 random dictonaries')
print('\n\n\nsegmentUid dictonary')

```
